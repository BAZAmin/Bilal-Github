```python
df["pickup_datetime"] = pd.to_datetime(df["pickup_datetime"], errors = "coerce")
```

```python
df = df [["pickup_datetime", "passenger_count"]]
```

```python
df.set_index("pickup_datetime", inplace = True)
```

```python
df = df.resample("D")["passenger_count"].sum().reset_index()
```

```python
data = df ["passenger_count"].values
data
```

```python
data
```

```python
sequence_length = 4
sequences = []
targets = []

for i in range(len(data)-sequence_length):
    sequences.append(data[i: i+sequence_length])
    targets.append(data[i+sequence_length])
```

```python
import numpy as np
sequences = np.array(sequences)
targets = np.array(targets)
```

```python
targets
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split (sequences, targets, test_size = 0.15)
```

```python
from tensorflow.keras import models, layers
```

```python
model = models.Sequential(name = "RNN")
model.add(layers.SimpleRNN(32, activation= "relu", input_shape = (X_train.shape[1], 1)))
model.add(layers. Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer = "adam", loss = "mean_squared_error")
model.fit(X_train, y_train, epochs = 500, verbose = 2)
```

```python
X_train.shape
```

```python
model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
y_pred = model.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
y_test, y_pred
```

```python
model = models.Sequential(name = "LSTM")
model.add(layers.LSTM(32, activation= "relu", input_shape = (X_train.shape[1], 1)))
model.add(layers. Dense(1))
model.compile(optimizer = "adam", loss = "mean_squared_error")
```
